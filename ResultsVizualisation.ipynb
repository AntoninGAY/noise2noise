```python
import imageio
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

list_images = [25, 50, 100]
nb_iter = 9000
folder = "results/00035-autoencoder'-test'-n2n/"
log = False

for img_nb in list_images:
    file_orig = folder + 'img_0_val_' + str(img_nb) + '_orig.png'
    file_nois = folder + 'img_0_val_' + str(img_nb) + '_noisy.png'
    file_pred = folder + 'img_' + str(nb_iter) + '_val_' + str(img_nb) + '_pred.png'
    
    im_orig = imageio.imread(file_orig)
    im_nois = imageio.imread(file_nois)
    im_pred = imageio.imread(file_pred)
    
    if log:
        im_orig = np.exp(im_orig)
        im_nois = np.exp(im_nois)
        im_pred = np.exp(im_pred)
        
    im_orig = im_orig.astype(np.int8)
    im_nois = im_nois.astype(np.int8)
    im_pred = im_pred.astype(np.int8)
    
    print("Shapes = %s | %s | %s" % (im_orig.shape, im_nois.shape, im_pred.shape))
        
    plt.figure()
    plt.subplot(131)
    plt.imshow(im_orig, cmap='gray')
    plt.title('Original Image')
    plt.subplot(132)
    plt.imshow(im_nois, cmap='gray')
    plt.title('Noisy Image')
    plt.subplot(133)
    plt.imshow(im_pred, cmap='gray')
    plt.title('Prediction Image')
    plt.show()
```
